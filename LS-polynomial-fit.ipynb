```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0,1,1000)
epsilon = np.random.normal(0,0.25,1000)

y = x + epsilon
```

```python
plt.scatter(x,y, c='purple', s=3)
plt.title('Dataset 2 Scatterplot')
plt.show()
```

```python
import scipy as sp

def f(a):
    return (x*a-y)**2

a_min = sp.optimize.leastsq(np.sum(f),x0=1)
print(a_min)

u=np.linspace(0,1,100)
plt.title('Least Squares Fit')
plt.plot(u,a_min[0]*u, c='green', label='Linear Function Fit')
plt.scatter(x,y,c='purple',s=3,label='Datapoints 1')
plt.legend(loc='upper right')
plt.show()
```

```python
x = np.random.uniform(0,1,1000)
ksi = np.random.normal(0,0.01,1000)

y = 30*((x-0.25)**2)*((x-0.75)**2) + ksi

plt.scatter(x,y,c='magenta',s=2)
plt.title('Dataset 2 Scatterplot')
plt.show()
```

```python
def p(x,p0,p1,p2,p3,p4):
    return p0+(p1*x)+(p2 * x**2)+(p3 * x**3)+(p4 * x**4)

yfit,_ = sp.optimize.curve_fit(p,x,y)
xplot = np.linspace(0,1,100)
yplot = p(xplot,yfit[0],yfit[1],yfit[2],yfit[3],yfit[4])
plt.scatter(x,y,c='magenta',s=2,label='Datapoints')
plt.plot(xplot,yplot,c='blue',label='Curve Fit')
plt.title('Polynomial Fit')
plt.legend(loc='upper right')
plt.show()
```
